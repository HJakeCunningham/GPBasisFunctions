```python
import numpy as np
from scipy.sparse import diags
from scipy import linalg
from scipy.sparse.linalg import inv
import matplotlib.pyplot as plt
```

## Symmetric band matrix

```python
n = 100
data = [np.ones(n-2), 2*np.ones(n-1), 3*np.ones(n), 2*np.ones(n-1), np.ones(n-2)]
offset = [-2,-1,0,1,2]
A_sparse = diags(data, offset)
A_dense = A_sparse.toarray()
```

```python
A_sparse_inv = inv(A_sparse).toarray()
plt.imshow(A_sparse_inv)
```

```python
A_dense_inv = linalg.inv(A_dense)
plt.imshow(A_dense_inv)
```
